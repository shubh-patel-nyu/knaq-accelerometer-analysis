```python
# ============================================================
# IMPORTS
# Load all libraries needed for the entire analysis
# pandas: handles tabular data (like Excel but in code)
# numpy: fast math and array operations
# matplotlib: draws all our graphs
# scipy: advanced signal processing (used for FFT)
# os: lets Python interact with files and folders
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import gzip
import os

print("All libraries loaded successfully")
```

```python
# ============================================================
# DATA LOADING FUNCTION
# This function reads a raw .log accelerometer file
# and separates it into valid data rows and corrupt rows
#
# Challenge: some files have a text header at the top,
# some start directly with data. This function handles both
# by detecting the first line that successfully parses as
# 4 numeric values — that's where data begins
# ============================================================

def load_log_file(filepath):
    rows = []         # stores valid data rows
    corrupt = []      # stores malformed/unreadable rows
    header_lines = [] # stores header text before data starts
    data_started = False  # tracks whether we've passed the header

    with open(filepath, 'r', errors='replace') as f:
        for i, line in enumerate(f):
            line = line.strip()

            if not data_started:
                # Try splitting line into 4 parts and converting to floats
                # If successful — data section has started
                # If not — it's still a header line, save and skip
                parts = line.split(',')
                if len(parts) == 4:
                    try:
                        float(parts[0]), float(parts[1]), float(parts[2]), float(parts[3])
                        data_started = True
                        # Don't skip — fall through and process this line as data
                    except ValueError:
                        header_lines.append(line)
                        continue
                else:
                    header_lines.append(line)
                    continue

            # Skip blank lines within the data section
            if line == '':
                continue

            # Every valid data line must have exactly 4 comma-separated values
            parts = line.split(',')
            if len(parts) != 4:
                corrupt.append({
                    'line_number': i+1,
                    'content': line,
                    'reason': 'wrong column count'
                })
                continue

            # Try converting all 4 values to floats
            # If any value is non-numeric, mark the entire row as corrupt
            try:
                x, y, z, t = float(parts[0]), float(parts[1]), float(parts[2]), float(parts[3])
                rows.append({'accel_x': x, 'accel_y': y, 'accel_z': z, 'timestamp': t})
            except ValueError:
                corrupt.append({
                    'line_number': i+1,
                    'content': line,
                    'reason': 'non-numeric value'
                })

    # Convert lists into pandas DataFrames for easy analysis
    df = pd.DataFrame(rows)
    corrupt_df = pd.DataFrame(corrupt)
    return df, corrupt_df, header_lines

# Load the October 2025 file as our representative dataset
# This file contains 18.28 hours of 800Hz accelerometer data
# chosen because it covers a full operational cycle including shutdown
df, corrupt, header = load_log_file('iov.log_2025_10_26.log')

print("Rows loaded:", len(df))
print("Corrupt rows:", len(corrupt))
```

```python
# ============================================================
# DATA QUALITY SUMMARY
# Convert raw unix timestamps to human-readable datetime
# Then calculate and print the full data quality report
# ============================================================

# Convert unix timestamp (seconds since 1970) to readable datetime
df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')

# Calculate time span of the recording
start_time = df['datetime'].min()
end_time = df['datetime'].max()
duration_seconds = df['timestamp'].max() - df['timestamp'].min()

# Effective sample rate = total valid samples divided by total time
# Should match the expected 800 Hz if data is complete
effective_rate = len(df) / duration_seconds

print("===== DATA QUALITY SUMMARY =====")
print(f"Total rows (valid):    {len(df):,}")
print(f"Corrupt rows:          {len(corrupt):,}")
print(f"Total rows processed:  {len(df) + len(corrupt):,}")
print(f"Corruption rate:       {len(corrupt)/(len(df)+len(corrupt))*100:.4f}%")
print(f"Start time:            {start_time}")
print(f"End time:              {end_time}")
print(f"Duration:              {duration_seconds/3600:.2f} hours")
print(f"Effective sample rate: {effective_rate:.1f} Hz")
print(f"Expected sample rate:  800 Hz")

# Show what the corrupt rows look like
# Finding: all corrupt rows are single '.' characters — a systematic
# heartbeat signal from the acquisition pipeline, not random corruption
print("\n===== SAMPLE CORRUPT ROWS =====")
print(corrupt.head(10).to_string())
```

```python
# ============================================================
# BASIC STATISTICS
# Compute mean, std, min, max, quartiles for each axis
# Also compute skewness (symmetry) and kurtosis (tail weight)
#
# Skewness near 0 = symmetric distribution (no persistent bias)
# Kurtosis ~1.4 = slightly lighter tails than normal distribution
# ============================================================

print("===== BASIC STATISTICS PER AXIS =====\n")
stats = df[['accel_x', 'accel_y', 'accel_z']].describe()
print(stats.to_string())

print("\n===== ADDITIONAL STATS =====")
for axis in ['accel_x', 'accel_y', 'accel_z']:
    skew = df[axis].skew()
    kurt = df[axis].kurtosis()
    print(f"{axis} — skewness: {skew:.4f}, kurtosis: {kurt:.4f}")
```

```python
# ============================================================
# ACCELERATION MAGNITUDE + PLOT SAMPLE
#
# Magnitude = sqrt(x² + y² + z²)
# This combines all three axes into a single scalar value
# representing total acceleration at each moment
#
# At rest: magnitude ≈ 9.81 m/s² (gravity only)
# During operation: magnitude > 9.81 due to added vibration
#
# We also create df_plot — a downsampled version for plotting
# Taking every 800th row gives 1 point per second
# (52 million points cannot be plotted directly)
# ============================================================

# Compute vector magnitude for every sample
df['magnitude'] = np.sqrt(df['accel_x']**2 + df['accel_y']**2 + df['accel_z']**2)

print("Magnitude stats:")
print(f"  Mean:  {df['magnitude'].mean():.4f} m/s²")
print(f"  Std:   {df['magnitude'].std():.4f} m/s²")
print(f"  Min:   {df['magnitude'].min():.4f} m/s²")
print(f"  Max:   {df['magnitude'].max():.4f} m/s²")

# Downsample for plotting: take 1 sample per second (every 800 rows)
df_plot = df.iloc[::800].copy().reset_index(drop=True)
print(f"\nPlot sample size: {len(df_plot):,} points (1 per second)")
```

```python
# ============================================================
# TIME DOMAIN PLOT
# Plot all three axes and magnitude over the full 18 hours
# sharex=True links all subplots to the same x-axis
# so zooming one zooms all
# ============================================================

fig, axes = plt.subplots(4, 1, figsize=(15, 12), sharex=True)

# Convert timestamp to hours from start for readable x-axis
time_hours = (df_plot['timestamp'] - df_plot['timestamp'].iloc[0]) / 3600

axes[0].plot(time_hours, df_plot['accel_x'], color='steelblue', linewidth=0.5)
axes[0].set_ylabel('Accel X (m/s²)')
axes[0].set_title('Acceleration X Axis over Time')
axes[0].grid(True, alpha=0.3)

axes[1].plot(time_hours, df_plot['accel_y'], color='darkorange', linewidth=0.5)
axes[1].set_ylabel('Accel Y (m/s²)')
axes[1].set_title('Acceleration Y Axis over Time')
axes[1].grid(True, alpha=0.3)

axes[2].plot(time_hours, df_plot['accel_z'], color='green', linewidth=0.5)
axes[2].set_ylabel('Accel Z (m/s²)')
axes[2].set_title('Acceleration Z Axis over Time')
axes[2].grid(True, alpha=0.3)

axes[3].plot(time_hours, df_plot['magnitude'], color='red', linewidth=0.5)
axes[3].set_ylabel('Magnitude (m/s²)')
axes[3].set_title('Acceleration Magnitude over Time')
axes[3].grid(True, alpha=0.3)

axes[3].set_xlabel('Time (hours from start)')
plt.suptitle('Time Domain Analysis — October 25-26 2025', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('time_domain.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved as time_domain.png")
```

```python
# ============================================================
# STATIONARY PERIOD DETECTION
#
# Key idea: when a machine is running, its acceleration varies
# a lot from moment to moment (high std). When stationary,
# it barely moves (very low std — just sensor noise)
#
# We use rolling standard deviation on magnitude with a
# 1-second window (800 samples) to measure local variability
#
# Threshold of 0.5 m/s² was chosen based on the data's
# natural bimodal distribution:
#   - Stationary: std clusters around ~0.02 m/s²
#   - Active: std clusters around ~1.2 m/s²
#   - Almost nothing exists between 0.03 and 1.1 m/s²
# ============================================================

# Rolling std and mean — step=800 means non-overlapping 1-second windows
rolling_std = df['magnitude'].rolling(window=800, step=800).std()
rolling_mean = df['magnitude'].rolling(window=800, step=800).mean()

# Build matching time axis (one point per second)
time_axis = df['timestamp'].iloc[::800].values
time_hours_full = (time_axis - time_axis[0]) / 3600

# Trim arrays to same length (rolling can produce slightly different lengths)
min_len = min(len(time_hours_full), len(rolling_std.dropna()))
time_hours_full = time_hours_full[:min_len]
rolling_std_vals = rolling_std.dropna().values[:min_len]
rolling_mean_vals = rolling_mean.dropna().values[:min_len]

# Apply threshold — True means stationary, False means active
STATIONARY_THRESHOLD = 0.5
stationary_mask = rolling_std_vals < STATIONARY_THRESHOLD

print(f"Stationary threshold: std < {STATIONARY_THRESHOLD} m/s²")
print(f"Total seconds analyzed: {min_len}")
print(f"Stationary seconds: {stationary_mask.sum():,}")
print(f"Active seconds: {(~stationary_mask).sum():,}")
print(f"Stationary percentage: {stationary_mask.sum()/min_len*100:.1f}%")

stationary_indices = np.where(stationary_mask)[0]
if len(stationary_indices) > 0:
    first_stationary = time_hours_full[stationary_indices[0]]
    print(f"\nMachine first goes stationary at: {first_stationary:.2f} hours from start")
```

```python
# ============================================================
# STATIONARY DETECTION EXPLORATION PLOT
# Plot rolling mean and rolling std to visually verify
# that our threshold correctly separates active from stationary
# The clear drop in std after ~12.5 hours confirms shutdown
# ============================================================

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8), sharex=True)

ax1.plot(time_hours_full, rolling_mean_vals, color='red', linewidth=0.5)
ax1.set_ylabel('Rolling Mean Magnitude (m/s²)')
ax1.set_title('Rolling Mean of Magnitude (1 second windows)')
ax1.grid(True, alpha=0.3)

ax2.plot(time_hours_full, rolling_std_vals, color='purple', linewidth=0.5)
ax2.axhline(y=0.5, color='orange', linestyle='--', label='threshold=0.5')
ax2.axhline(y=1.0, color='red', linestyle='--', label='threshold=1.0')
ax2.set_ylabel('Rolling Std of Magnitude (m/s²)')
ax2.set_title('Rolling Std of Magnitude — used to detect stationary periods')
ax2.set_xlabel('Time (hours from start)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.suptitle('Stationary Period Detection', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('stationary_detection.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# ACTIVE VS STATIONARY FINAL PLOT + PERIOD LIST
#
# Green shading = stationary (std < 0.5)
# Red shading = active (std >= 0.5)
#
# After plotting, we find the raw stationary periods
# (before merging — shows all individual quiet moments)
# ============================================================

STATIONARY_THRESHOLD = 0.5
stationary_mask = rolling_std_vals < STATIONARY_THRESHOLD

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(time_hours_full, rolling_std_vals, color='purple', linewidth=0.5, label='Rolling Std')
ax.axhline(y=STATIONARY_THRESHOLD, color='red', linestyle='--',
           linewidth=1.5, label=f'Threshold = {STATIONARY_THRESHOLD}')

# Shade regions green (stationary) and red (active)
ax.fill_between(time_hours_full, 0, rolling_std_vals,
                where=stationary_mask, color='lightgreen', alpha=0.5, label='Stationary')
ax.fill_between(time_hours_full, 0, rolling_std_vals,
                where=~stationary_mask, color='lightsalmon', alpha=0.5, label='Active')

ax.set_xlabel('Time (hours from start)')
ax.set_ylabel('Rolling Std of Magnitude (m/s²)')
ax.set_title('Machine Active vs Stationary Periods')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('active_vs_stationary.png', dpi=150, bbox_inches='tight')
plt.show()

# Detect transitions between active and stationary
# np.diff finds where the mask changes (0→1 = goes stationary, 1→0 = goes active)
changes = np.diff(stationary_mask.astype(int))
starts = np.where(changes == 1)[0]
ends = np.where(changes == -1)[0]

print(f"\nDetected stationary periods:")
for i, (s, e) in enumerate(zip(starts, ends)):
    duration = time_hours_full[e] - time_hours_full[s]
    print(f"  Period {i+1}: {time_hours_full[s]:.2f}h to {time_hours_full[e]:.2f}h — duration: {duration*60:.1f} minutes")
```

```python
# ============================================================
# MERGE STATIONARY PERIODS
#
# Raw detection gives many tiny periods (some just seconds long)
# because the machine briefly quiets mid-operation
#
# We merge periods that are within 60 seconds of each other
# and filter out anything shorter than 30 seconds
# This gives us meaningful operational pauses and shutdowns
# ============================================================

MIN_DURATION_SECONDS = 30  # ignore periods shorter than 30 seconds
MERGE_GAP_SECONDS = 60     # merge periods within 60 seconds of each other

changes = np.diff(stationary_mask.astype(int))
starts_raw = np.where(changes == 1)[0]
ends_raw = np.where(changes == -1)[0]

# Handle edge cases where recording starts or ends in stationary state
if stationary_mask[0]:
    starts_raw = np.concatenate([[0], starts_raw])
if stationary_mask[-1]:
    ends_raw = np.concatenate([ends_raw, [len(stationary_mask)-1]])

# Merge nearby periods into single events
merged_starts = []
merged_ends = []
i = 0
while i < len(starts_raw):
    s = starts_raw[i]
    e = ends_raw[i]
    # If the next period starts within 60 seconds, absorb it into current period
    while i + 1 < len(starts_raw) and (starts_raw[i+1] - ends_raw[i]) < MERGE_GAP_SECONDS:
        i += 1
        e = ends_raw[i]
    merged_starts.append(s)
    merged_ends.append(e)
    i += 1

print("===== CLEANED STATIONARY PERIODS =====")
significant = []
for i, (s, e) in enumerate(zip(merged_starts, merged_ends)):
    duration_sec = (time_hours_full[e] - time_hours_full[s]) * 3600
    if duration_sec >= MIN_DURATION_SECONDS:
        significant.append((s, e, duration_sec))
        start_actual = df['datetime'].iloc[s*800]
        print(f"Period {len(significant)}: {time_hours_full[s]:.2f}h to {time_hours_full[e]:.2f}h "
              f"— duration: {duration_sec/60:.1f} minutes "
              f"— starts at {start_actual.strftime('%H:%M:%S')}")

print(f"\nTotal significant stationary periods: {len(significant)}")
```

```python
# ============================================================
# FREQUENCY DOMAIN ANALYSIS (FFT)
#
# FFT = Fast Fourier Transform
# Converts a time-domain signal into frequency components
# Answers the question: "what vibration frequencies does
# this machine operate at?"
#
# We use the first 10 minutes of active operation (480,000
# samples) to get a clean baseline frequency signature
#
# Key findings:
#   ~24 Hz = primary operating frequency (fundamental)
#   ~52 Hz = second harmonic (2x fundamental, common in rotating machinery)
#   ~8.22 Hz = lower structural resonance
# ============================================================

SAMPLE_RATE = 800  # Hz — samples per second
SEGMENT_SIZE = SAMPLE_RATE * 60 * 10  # 10 minutes = 480,000 samples

# Use first hour of data to ensure we're in active operation
df_active = df[df['datetime'] < df['datetime'].iloc[0] + pd.Timedelta(hours=1)].copy()
segment = df_active.iloc[:SEGMENT_SIZE]

print(f"FFT segment size: {len(segment):,} samples")
print(f"FFT segment duration: {len(segment)/SAMPLE_RATE/60:.1f} minutes")

fig, axes = plt.subplots(3, 1, figsize=(15, 10))

for idx, (axis, color) in enumerate(zip(['accel_x', 'accel_y', 'accel_z'],
                                         ['steelblue', 'darkorange', 'green'])):
    data = segment[axis].values

    # rfftfreq gives the frequency bins for real-valued FFT
    # d=1/SAMPLE_RATE sets the time spacing between samples
    freqs = np.fft.rfftfreq(len(data), d=1/SAMPLE_RATE)

    # rfft computes the FFT, abs gives magnitude, divide by length to normalize
    fft_vals = np.abs(np.fft.rfft(data)) / len(data)

    # Only plot up to 100 Hz — above this is noise for our machine type
    mask = freqs <= 100
    axes[idx].plot(freqs[mask], fft_vals[mask], color=color, linewidth=0.5)
    axes[idx].set_ylabel('Amplitude')
    axes[idx].set_title(f'Frequency Domain — {axis}')
    axes[idx].grid(True, alpha=0.3)

    # Find and print the 5 strongest frequency components
    top_idx = np.argsort(fft_vals[mask])[-5:][::-1]
    top_freqs = freqs[mask][top_idx]
    top_amps = fft_vals[mask][top_idx]
    print(f"\n{axis} top 5 dominant frequencies:")
    for f, a in zip(top_freqs, top_amps):
        print(f"  {f:.2f} Hz — amplitude: {a:.4f}")

axes[2].set_xlabel('Frequency (Hz)')
plt.suptitle('Frequency Domain Analysis (first 10 min of operation)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('frequency_domain.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# OUTLIER DETECTION USING Z-SCORE
#
# Z-score measures how many standard deviations a value is
# from the mean. Formula: z = (value - mean) / std
#
# A z-score of 5 means the value is 5 standard deviations
# from the mean — extremely unlikely in normal operation
#
# We apply this to magnitude to find the most extreme samples
# Threshold of ±5 was chosen to capture genuine anomalies
# while excluding normal operational variation
# ============================================================

Z_THRESHOLD = 5

# Compute z-score for every magnitude value
df['mag_zscore'] = (df['magnitude'] - df['magnitude'].mean()) / df['magnitude'].std()

# Extract all rows where magnitude is anomalously high or low
outliers = df[np.abs(df['mag_zscore']) > Z_THRESHOLD].copy()

print(f"Z-score threshold: ±{Z_THRESHOLD}")
print(f"Total outliers found: {len(outliers):,}")
print(f"Outlier percentage: {len(outliers)/len(df)*100:.4f}%")
print(f"Min magnitude in outliers: {outliers['magnitude'].min():.4f} m/s²")
print(f"Max magnitude in outliers: {outliers['magnitude'].max():.4f} m/s²")
print(f"Mean magnitude in outliers: {outliers['magnitude'].mean():.4f} m/s²")
```

```python
# ============================================================
# OUTLIER PLOT
#
# Plot magnitude over time with outlier points shown in red
# df_plot_trimmed fixes a 1-row length mismatch between
# df_plot (65,826 rows) and time_hours_full (65,825 rows)
# that would otherwise cause a plotting error
# ============================================================

# Trim to fix length mismatch
df_plot_trimmed = df_plot.iloc[:len(time_hours_full)].copy()

fig, ax = plt.subplots(figsize=(15, 5))

# Plot full magnitude signal in blue
ax.plot(time_hours_full, df_plot_trimmed['magnitude'],
        color='steelblue', linewidth=0.5, label='Magnitude', alpha=0.7)

# Overlay outlier points in red
outlier_times = (outliers['timestamp'] - df['timestamp'].iloc[0]) / 3600
ax.scatter(outlier_times, outliers['magnitude'],
           color='red', s=5, label=f'Outliers (z>{Z_THRESHOLD})', zorder=5)

ax.set_xlabel('Time (hours from start)')
ax.set_ylabel('Magnitude (m/s²)')
ax.set_title(f'Acceleration Magnitude with Outliers Highlighted (Z-score > {Z_THRESHOLD})')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('outliers.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# TRANSIENT EVENT DETECTION
#
# A transient event = a sudden spike above 20.0 m/s²
# This threshold is ~9.2 standard deviations above the mean
# making it extremely selective — only genuine shocks qualify
#
# Events within 800 samples (1 second) of each other are
# grouped together as a single event, since a shock often
# affects multiple consecutive samples
# ============================================================

TRANSIENT_THRESHOLD = 20.0  # m/s² — minimum magnitude to qualify as an event
MIN_GAP_SAMPLES = 800        # minimum gap between separate events (1 second)

# Find indices of all samples exceeding the threshold
above = df['magnitude'] > TRANSIENT_THRESHOLD
above_idx = np.where(above.values)[0]

print(f"Samples above {TRANSIENT_THRESHOLD} m/s²: {len(above_idx):,}")

# Group consecutive above-threshold samples into single events
events = []
if len(above_idx) > 0:
    event_start = above_idx[0]
    event_end = above_idx[0]

    for i in range(1, len(above_idx)):
        if above_idx[i] - above_idx[i-1] < MIN_GAP_SAMPLES:
            # Still within same event — extend the end
            event_end = above_idx[i]
        else:
            # Gap is large enough — save current event, start new one
            events.append((event_start, event_end))
            event_start = above_idx[i]
            event_end = above_idx[i]
    events.append((event_start, event_end))

print(f"Total transient events found: {len(events)}")
print(f"\n===== ALL TRANSIENT EVENTS =====")

# Print details for every detected event
for i, (s, e) in enumerate(events):
    duration_ms = (df['timestamp'].iloc[e] - df['timestamp'].iloc[s]) * 1000
    peak_mag = df['magnitude'].iloc[s:e+1].max()
    peak_x = df['accel_x'].iloc[s:e+1].abs().max()
    peak_y = df['accel_y'].iloc[s:e+1].abs().max()
    peak_z = df['accel_z'].iloc[s:e+1].abs().max()
    event_time = df['datetime'].iloc[s]
    time_hr = (df['timestamp'].iloc[s] - df['timestamp'].iloc[0]) / 3600

    print(f"\nEvent {i+1}:")
    print(f"  Time:        {event_time.strftime('%H:%M:%S')} ({time_hr:.3f} hrs from start)")
    print(f"  Duration:    {duration_ms:.1f} ms")
    print(f"  Peak mag:    {peak_mag:.4f} m/s²")
    print(f"  Peak |X|:    {peak_x:.4f} m/s²")
    print(f"  Peak |Y|:    {peak_y:.4f} m/s²")
    print(f"  Peak |Z|:    {peak_z:.4f} m/s²")
```

```python
# ============================================================
# TOP 10 TRANSIENT EVENTS — DETAILED CHARACTERIZATION
#
# From 311 total events, we hand-pick 10 based on:
#   - Peak magnitude (highest severity)
#   - Unusual duration patterns (235ms recurring events)
#   - Anomalous axis behavior (Y-axis suppression)
#
# For each event we plot 2 seconds of context around it
# so we can see the signal before, during, and after the spike
# ============================================================

# Selected event indices (0-based) chosen for maximum insight
selected = [0, 10, 95, 136, 140, 148, 216, 229, 230, 282]

print("===== TOP 10 CHARACTERIZED TRANSIENT EVENTS =====\n")

fig, axes = plt.subplots(10, 1, figsize=(15, 40))

for plot_idx, event_idx in enumerate(selected):
    s, e = events[event_idx]

    # Extract 2 seconds of data before and after the event for context
    context_start = max(0, s - 1600)   # 1600 samples = 2 seconds at 800Hz
    context_end = min(len(df)-1, e + 1600)
    context = df.iloc[context_start:context_end].copy()

    # Time axis in milliseconds relative to event start (0 = event begins)
    context_time_ms = (context['timestamp'] - df['timestamp'].iloc[s]) * 1000

    # Calculate event metrics
    duration_ms = (df['timestamp'].iloc[e] - df['timestamp'].iloc[s]) * 1000
    peak_mag = df['magnitude'].iloc[s:e+1].max()
    peak_x = df['accel_x'].iloc[s:e+1].abs().max()
    peak_y = df['accel_y'].iloc[s:e+1].abs().max()
    peak_z = df['accel_z'].iloc[s:e+1].abs().max()
    event_time = df['datetime'].iloc[s]
    time_hr = (df['timestamp'].iloc[s] - df['timestamp'].iloc[0]) / 3600

    # Find which axis had the highest peak
    axis_peaks = {'X': peak_x, 'Y': peak_y, 'Z': peak_z}
    dominant = max(axis_peaks, key=axis_peaks.get)

    print(f"Event {plot_idx+1} (raw event #{event_idx+1}):")
    print(f"  Time:           {event_time.strftime('%H:%M:%S')} ({time_hr:.3f} hrs)")
    print(f"  Duration:       {duration_ms:.1f} ms")
    print(f"  Peak magnitude: {peak_mag:.4f} m/s²")
    print(f"  Peak |X|:       {peak_x:.4f} m/s²")
    print(f"  Peak |Y|:       {peak_y:.4f} m/s²")
    print(f"  Peak |Z|:       {peak_z:.4f} m/s²")
    print(f"  Dominant axis:  {dominant}")
    print()

    # Plot magnitude and Z axis for each event
    # Vertical dashed line marks exact moment of event start
    axes[plot_idx].plot(context_time_ms, context['magnitude'],
                        color='red', linewidth=0.8, label='Magnitude')
    axes[plot_idx].plot(context_time_ms, context['accel_z'],
                        color='green', linewidth=0.5, alpha=0.7, label='Z')
    axes[plot_idx].axvline(x=0, color='black', linestyle='--', linewidth=1, label='Event start')
    axes[plot_idx].set_title(
        f'Event {plot_idx+1} — {event_time.strftime("%H:%M:%S")} | '
        f'Peak: {peak_mag:.2f} m/s² | Duration: {duration_ms:.1f}ms | Dominant: {dominant}')
    axes[plot_idx].set_ylabel('m/s²')
    axes[plot_idx].grid(True, alpha=0.3)
    axes[plot_idx].legend(loc='upper right', fontsize=7)

axes[-1].set_xlabel('Time relative to event (ms)')
plt.suptitle('Top 10 Transient Events — Detailed View', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('top10_events.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved.")
```

```python
# ============================================================
# EVENT COMPARISON AND ANOMALY ANALYSIS
#
# Build a summary DataFrame of the top 10 events
# Then compare them against each other to find:
#   1. Duration clusters (what patterns repeat?)
#   2. Y-axis anomalies (which events have suppressed Y?)
#   3. Magnitude outliers (which are most severe?)
#   4. Temporal clustering (do events happen in bursts?)
# ============================================================

event_summary = {
    'Event': list(range(1, 11)),
    'Time': ['12:27:38','12:59:22','15:26:05','17:38:40','17:46:36',
             '17:59:25','19:41:23','19:54:19','19:56:42','22:06:38'],
    'Hours_from_start': [0.455,0.984,3.430,5.639,5.772,5.985,7.685,7.900,7.940,10.106],
    'Duration_ms': [0.0,235.0,2.5,228.7,235.0,0.0,0.0,2.5,235.0,0.0],
    'Peak_mag': [23.29,23.36,29.35,22.65,24.01,26.32,26.06,26.94,22.77,29.11],
    'Peak_X': [13.49,13.84,12.16,12.77,13.16,11.40,12.00,12.22,14.08,10.66],
    'Peak_Y': [5.47,7.44,6.49,6.58,5.88,0.05,1.94,5.93,6.13,2.65],
    'Peak_Z': [18.18,21.20,26.06,19.48,19.98,23.73,23.05,23.27,19.24,26.95]
}

df_events = pd.DataFrame(event_summary)

print("===== EVENT COMPARISON =====\n")
print(df_events.to_string(index=False))

print("\n===== ANOMALIES AMONG THE 10 EVENTS =====\n")

# --- Duration clustering ---
# Three distinct clusters suggest three different types of mechanical events
print("Duration patterns:")
print(f"  Zero-duration events (single sample): {len(df_events[df_events.Duration_ms == 0])}")
print(f"  ~235ms duration events: {len(df_events[df_events.Duration_ms > 200])}")
print(f"  ~2.5ms duration events: {len(df_events[(df_events.Duration_ms > 0) & (df_events.Duration_ms < 10)])}")

# --- Y-axis anomaly detection ---
# Low Y means impact energy is concentrated in XZ plane
# This is structurally different from normal shock behavior
print("\nY-axis anomalies:")
y_mean = df_events['Peak_Y'].mean()
y_std = df_events['Peak_Y'].std()
for _, row in df_events.iterrows():
    zscore = (row['Peak_Y'] - y_mean) / y_std
    if abs(zscore) > 1.5:
        print(f"  Event {int(row['Event'])}: Peak Y = {row['Peak_Y']:.2f} m/s² (z-score: {zscore:.2f}) — ANOMALOUS")

# --- Magnitude outliers within the top 10 ---
print("\nMagnitude anomalies:")
mag_mean = df_events['Peak_mag'].mean()
mag_std = df_events['Peak_mag'].std()
for _, row in df_events.iterrows():
    zscore = (row['Peak_mag'] - mag_mean) / mag_std
    if abs(zscore) > 1.0:
        print(f"  Event {int(row['Event'])}: Peak mag = {row['Peak_mag']:.2f} m/s² (z-score: {zscore:.2f})")

# --- Temporal clustering ---
# Events close together in time suggest operational stress phases
print("\nTemporal clustering:")
for i in range(1, len(df_events)):
    gap = df_events['Hours_from_start'].iloc[i] - df_events['Hours_from_start'].iloc[i-1]
    if gap < 0.5:
        print(f"  Events {i} and {i+1} are close: gap = {gap*60:.1f} minutes")
```

```python
# ============================================================
# FINAL SUMMARY VISUALIZATION (3 panels)
#
# Panel 1: Full magnitude timeline with top 10 events marked
#   Color coding: darkred = most severe (>27), red = severe (>25),
#   orange = moderate
#
# Panel 2: Scatter plot of severity vs time
#   Dot size = event duration (small=instant, large=235ms)
#   Dot color = Z-axis intensity (hotter = more Z energy)
#
# Panel 3: Bar chart of Y-axis peaks per event
#   Red bars = anomalously low Y (below 2.0 m/s²)
#   These events have a fundamentally different shock signature
# ============================================================

fig, axes = plt.subplots(3, 1, figsize=(15, 14))

# --- Panel 1: Timeline with events marked ---
axes[0].plot(time_hours_full, df_plot_trimmed['magnitude'],
             color='steelblue', linewidth=0.5, alpha=0.8, label='Magnitude')
axes[0].set_ylim(0, 32)

for i, row in df_events.iterrows():
    color = 'darkred' if row['Peak_mag'] > 27 else 'red' if row['Peak_mag'] > 25 else 'orange'
    axes[0].axvline(x=row['Hours_from_start'], color=color, linewidth=1.5, alpha=0.8)
    axes[0].text(row['Hours_from_start'] + 0.05, 30, f"E{int(row['Event'])}",
                fontsize=8, color=color, fontweight='bold')

axes[0].set_ylabel('Magnitude (m/s²)')
axes[0].set_xlabel('Time (hours from start)')
axes[0].set_title('Acceleration Magnitude with Top 10 Events Marked')
axes[0].grid(True, alpha=0.3)

# --- Panel 2: Severity scatter ---
# Dot size encodes duration: 60=instant, 150=brief, 400=sustained
scatter_sizes = [60 if d == 0 else 150 if d < 10 else 400 for d in df_events['Duration_ms']]
sc = axes[1].scatter(df_events['Hours_from_start'], df_events['Peak_mag'],
                     s=scatter_sizes, c=df_events['Peak_Z'],
                     cmap='YlOrRd', zorder=5, edgecolors='black', linewidth=0.8,
                     vmin=17, vmax=27)
plt.colorbar(sc, ax=axes[1], label='Peak Z Acceleration (m/s²)')

for _, row in df_events.iterrows():
    axes[1].annotate(f"E{int(row['Event'])}",
                    (row['Hours_from_start'], row['Peak_mag']),
                    textcoords='offset points', xytext=(5, 5), fontsize=9)

axes[1].set_ylim(20, 32)
axes[1].set_ylabel('Peak Magnitude (m/s²)')
axes[1].set_xlabel('Time (hours from start)')
axes[1].set_title('Event Severity vs Time (dot size = duration: small=instant, large=~235ms)')
axes[1].grid(True, alpha=0.3)

# --- Panel 3: Y-axis suppression bar chart ---
event_nums = [int(x) for x in df_events['Event'].tolist()]
y_vals = df_events['Peak_Y'].tolist()
bar_colors = ['red' if y < 2.0 else 'steelblue' for y in y_vals]

axes[2].bar(event_nums, y_vals, color=bar_colors, edgecolor='black', linewidth=0.8, width=0.6)
axes[2].axhline(y=2.0, color='red', linestyle='--', linewidth=1.5, label='Anomaly threshold (Y < 2)')

# Add value labels on top of each bar
for i, (num, val) in enumerate(zip(event_nums, y_vals)):
    axes[2].text(num, val + 0.1, f'{val:.2f}', ha='center', fontsize=8)

axes[2].set_xticks(event_nums)
axes[2].set_xticklabels([f'E{n}' for n in event_nums])
axes[2].set_xlabel('Event')
axes[2].set_ylabel('Peak |Y| Acceleration (m/s²)')
axes[2].set_title('Y-Axis Peak per Event (red = anomalously low, suggesting pure XZ-plane impact)')
axes[2].set_ylim(0, 9)
axes[2].legend()
axes[2].grid(True, alpha=0.3, axis='y')

plt.suptitle('Transient Event Analysis Summary', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('event_summary.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved.")
```
